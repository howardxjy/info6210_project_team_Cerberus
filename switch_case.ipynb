```python
sql_6 = """SELECT Resort_Name , State , Adult_ticket FROM skiresort_usa WHERE Adult_ticket >= 120 AND State LIKE %s"""
sql_7 = """SELECT Resort_Name , State , Adult_ticket FROM skiresort_usa WHERE State LIKE %s AND Adult_ticket >= 60 AND Adult_ticket < 120"""
sql_8 = """SELECT Resort_Name , State , Adult_ticket FROM skiresort_usa WHERE Adult_ticket < 60 AND State LIKE %s"""
```

```python
#get the associated ski resorts in a given state around the given adult ticket price
def f5():
    d = input('please input your state : ')
    a = input('please input your expected ticket price : ')
    b = int(a)
    if b >= 120:
        c.execute(sql_6, d)
        for row in c.fetchall():
            print(row,'expensive');
    elif 60 <= b < 120:
        c.execute(sql_7, d)
        for row in c.fetchall():
            print(row,'medium') ;
    elif b < 60:
        c.execute(sql_8, d)
        for row in c.fetchall():
            print(row,'cheap')  ;       
        
```

```python
f5()
```

```python
sql_7 = """SELECT twitter_name , twitter_screen_name , location , followers_count FROM project_twitter WHERE followers_count < 3000"""
sql_8 = """SELECT twitter_name , twitter_screen_name , location , followers_count FROM project_twitter WHERE followers_count >= 3000 AND followers_count <10000"""
sql_9 = """SELECT twitter_name , twitter_screen_name , location , followers_count FROM project_twitter WHERE followers_count >= 10000 AND followers_count < 30000"""
sql_10 = """SELECT twitter_name , twitter_screen_name , location , followers_count FROM project_twitter WHERE followers_count >= 30000"""
```

```python
#find a twitter account is famous or not
def f6():
    rep = input("Please input the reputation : unfamous, little_famous, state_famous, very_famous")
    re = str(rep)
    if re in 'unfamous':
        c.execute(sql_7)
        for row in c.fetchall():
            print(row,rep);
    elif re in 'little_famous':
        c.execute(sql_8)
        for row in c.fetchall():
            print(row,rep);
    elif re in 'state_famous':
        c.execute(sql_9)
        for row in c.fetchall():
            print(row,rep);
    elif re in 'very_famous':
        c,execute(sql_10)
        for row in c.fetchall():
            print(row,rep);
    
```
